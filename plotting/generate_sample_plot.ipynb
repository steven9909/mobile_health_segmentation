```python
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
labels = ["w/o", "w"]
group1 = [20, 35, 30, 28]
group2 = [5, 10, 6, 7]

bar_width = 0.25

r1 = np.arange(len(group1))
r2 = [x + bar_width for x in r1]

# Make the plot
plt.bar(r1, group1, color='b', width=bar_width, edgecolor='grey', label='Without Software Assistance')
plt.bar(r2, group2, color='r', width=bar_width, edgecolor='grey', label='With Software Assistance')

plt.xlabel('Participants', fontweight='bold', fontsize=15)
plt.xticks([r + bar_width / 2 for r in range(len(group1))], ['P1', 'P2', 'P3', 'Mean'])
#plt.xticks([r + l * bar_width for r in range(len(group1)) for l in range(len(labels))], labels * 4)
plt.ylabel("Absolute Difference", fontweight='bold', fontsize=15)

def add_significance_bar(start, end, height_left, height_right, star):
    plt.plot([start, start, end, end], [height_left + 0.5, max(height_left, height_right) + 1, max(height_left, height_right) + 1, height_right + 0.5], lw=1.5, c='k')
    plt.text((start + end) * 0.5, max(height_right, height_left) + 1, star, ha='center', va='bottom', fontsize=12)

def add_error_bar(start, height_low, height_high):
    s = start - 0.05
    e = start + 0.05
    plt.plot([s, e, (s + e) / 2, (s + e) / 2, s, e], [height_low, height_low, height_low, height_high, height_high, height_high], lw=1.5, c='k')

add_significance_bar(0, 0.25, 23, 8, "*")
add_significance_bar(1, 1.25, 39, 14, "**")
add_significance_bar(2, 2.25, 35, 8, "***")
add_significance_bar(3, 3.25, 34, 10, "**")
add_error_bar(0, 17, 22)
add_error_bar(0.25, 2, 7)
add_error_bar(1, 30, 38)
add_error_bar(1.25, 5, 13)
add_error_bar(2, 25, 34)
add_error_bar(2.25, 2, 7)
add_error_bar(3, 25, 33)
add_error_bar(3.25, 4, 9)

plt.ylim(0, 50)
plt.legend()
os.makedirs("./figures", exist_ok=True)
plt.savefig("./figures/plot.png", dpi=300, bbox_inches="tight")
```

```python
plt.figure()
plt.scatter([np.random.normal() for _ in range(100)], [np.random.normal() for _ in range(100)])
plt.xlabel("Average Measurements")
plt.ylabel("Difference between Measurements")
plt.xlim(-3, 3)
plt.ylim(-3, 3)

plt.plot([-3, 3], [-2, -2], c='b', lw=1.5, linestyle='--')
plt.plot([-3, 3], [1.7, 1.7], c='r', lw=1.5, linestyle='--')
plt.plot([-3, 3], [-0.15, -0.15], c='k', lw=1.5)

plt.text(3, 1.72, "+2.98SD", ha='right', va='bottom', fontsize=8, c='r')
plt.text(3, 1.5, "+1.7", ha='right', va='bottom', fontsize=8, c='r')

plt.text(3, -1.98, "+1.75SD", ha='right', va='bottom', fontsize=8, c='b')
plt.text(3, -2.18, "-2", ha='right', va='bottom', fontsize=8, c='b')

plt.text(3, -0.12, "MEAN", ha='right', va='bottom', fontsize=8)
plt.text(3, -0.35, "-0.15", ha='right', va='bottom', fontsize=8)

plt.title("Bland-Altman Plot")
plt.savefig("./figures/bland_altman.png", dpi=300, bbox_inches="tight")
```
